```python
import csv
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
```

```python
lines = list(range(0,1200))
magnitude = []
placename = []
time = []
updated = []
tz = []
urllog = []
details = []
felt = []
cdi = []
mmi = []
alert = []
status = []
tsunami = []
sig = []
net = []
code = []
ids = []
sources = []
types = []
nst = []
dmin = []
rms = []
gap = []
magtype = []
etype = []
title = []
ltype = []
coord = []
```

```python
url = "https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&"
start = "starttime=1960-01-01"
end = "endtime=2020-03-02"
lat = "latitude=64.9311916"
lng = "longitude=-19.0211694"
radius = "maxradiuskm=500"

searchurl = f"{url}{start}&{end}&{lat}&{lng}&{radius}"
response = requests.get(searchurl).json()
print(searchurl)
#print(response)
# test = response["features"][0]
# print(test)

for row in lines:
    try:
        magnitude.append(response["features"][row]["properties"]["mag"])
        placename.append(response["features"][row]["properties"]["place"])
        time.append(response["features"][row]["properties"]["time"])
        updated.append(response["features"][row]["properties"]["updated"])
        tz.append(response["features"][row]["properties"]["tz"])
        urllog.append(response["features"][row]["properties"]["url"])
        details.append(response["features"][row]["properties"]["detail"])
        felt.append(response["features"][row]["properties"]["felt"])
        cdi.append(response["features"][row]["properties"]["cdi"])
        mmi.append(response["features"][row]["properties"]["mmi"])
        alert.append(response["features"][row]["properties"]["alert"])
        status.append(response["features"][row]["properties"]["status"])
        tsunami.append(response["features"][row]["properties"]["tsunami"])
        sig.append(response["features"][row]["properties"]["sig"])
        net.append(response["features"][row]["properties"]["net"])
        code.append(response["features"][row]["properties"]["code"])
        ids.append(response["features"][row]["properties"]["ids"])
        sources.append(response["features"][row]["properties"]["sources"])
        types.append(response["features"][row]["properties"]["types"])
        nst.append(response["features"][row]["properties"]["nst"])
        dmin.append(response["features"][row]["properties"]["dmin"])
        rms.append(response["features"][row]["properties"]["rms"])
        gap.append(response["features"][row]["properties"]["gap"])
        magtype.append(response["features"][row]["properties"]["magType"])
        etype.append(response["features"][row]["properties"]["type"])
        title.append(response["features"][row]["properties"]["title"])
        ltype.append(response["features"][row]["geometry"]["type"])
        coord.append(response["features"][row]["geometry"]["coordinates"])
    except:
        magnitude.append(0)
        placename.append(0)
        time.append(0)
        updated.append(0)
        tz.append(0)
        urllog.append(0)
        details.append(0)
        felt.append(0)
        cdi.append(0)
        mmi.append(0)
        alert.append(0)
        status.append(0)
        tsunami.append(0)
        sig.append(0)
        net.append(0)
        code.append(0)
        ids.append(0)
        sources.append(0)
        types.append(0)
        nst.append(0)
        dmin.append(0)
        rms.append(0)
        gap.append(0)
        magtype.append(0)
        etype.append(0)
        title.append(0)
        ltype.append(0)
        coord.append(0)
    
```

```python
earthquakedf = pd.DataFrame(lines)
earthquakedf
```

```python
earthquakedf["Magnitude"] = magnitude
earthquakedf["Name"] = placename
earthquakedf["Time"] = time
earthquakedf["Last Updated"] = updated
earthquakedf["TZ"] = tz
earthquakedf["URL"] = urllog
earthquakedf["Details"] = details
earthquakedf["Felt"] = felt
earthquakedf["CDI"] = cdi
earthquakedf["MMI"] = mmi
earthquakedf["Alert"] = alert
earthquakedf["Status"] = status
earthquakedf["Tsunami"] = tsunami
earthquakedf["SIG"] = sig
earthquakedf["Net"] = net
earthquakedf["Code"] = code
earthquakedf["IDS"] = ids
earthquakedf["Sources"] = sources
earthquakedf["Types"] = types
earthquakedf["NST"] = nst
earthquakedf["DMIN"] = dmin
earthquakedf["RMS"] = rms
earthquakedf["GAP"] = gap
earthquakedf["MagType"] = magtype
earthquakedf["Etype"] = etype
earthquakedf["Title"] = title
earthquakedf["Location Type"] = ltype
earthquakedf["Coordinates"] = coord
```

```python
earthquakedf = earthquakedf[earthquakedf.Magnitude != 0]
earthquakedf.to_csv("earthquakedata.csv",index=False)
earthquakedf
```
